# importing some important libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
df = pd.read_csv('Churn_Modelling.csv')
```

```python
df.head()
```

# Summary of the dataset

```python
df.info()
```

# Drop the Unnecessary Column

```python
df.drop(columns=['RowNumber', 'CustomerId', 'Surname'], axis = 1, inplace = True)
```

```python
df.head()
```

# one-hot encoding technique on categorical variables

```python
df = pd.get_dummies(df, columns = ['Geography', "Gender"], drop_first=True)
df.head()
```

```python
x = df.drop(columns = ['Exited'])
y = df['Exited']
```

```python
from sklearn.model_selection import train_test_split
# Correct the order of variables assigned from train_test_split
X_train, x_test, Y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
# Fit and transform only the training features
X_train = sc.fit_transform(X_train)
# Transform the test features using the scaler fitted on the training data
x_test = sc.transform(x_test)
# Do not scale the target variables Y_train and y_test
```

```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(11, activation='relu', input_dim=11))
model.add(Dense(11, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.summary()
```

```python
x.shape
```

```python
y.shape
```

```python
model.compile(optimizer='Adam',loss='binary_crossentropy',metrics=['accuracy'])
history = model.fit(X_train,Y_train,batch_size=50,epochs=100,verbose=1,validation_split=0.2)
```

```python
y_pred = model.predict(x_test)
y_pred
y_pred = y_pred.argmax(axis=-1)
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
import numpy as np

# Take user input for relevant features
print("Please provide the following details for prediction:")
credit_score = float(input("Credit Score: "))
age = float(input("Age: "))
tenure = float(input("Tenure: "))
balance = float(input("Balance: "))
num_of_products = float(input("Number of Products: "))
has_credit_card = float(input("Has Credit Card (1 for Yes, 0 for No): "))
is_active_member = float(input("Is Active Member (1 for Yes, 0 for No): "))
estimated_salary = float(input("Estimated Salary: "))
geography = input("Geography (France, Germany, Spain): ")
gender = input("Gender (Male, Female): ")

# Encode categorical variables
geography_mapping = {'France': [1, 0], 'Germany': [0, 1], 'Spain': [0, 0]}
gender_mapping = {'Male': [1], 'Female': [0]}
encoded_geography = geography_mapping.get(geography, [0, 0])
encoded_gender = gender_mapping.get(gender, [0])

# Combine all features into an input array
user_input = np.array([[credit_score, age, tenure, balance, num_of_products, has_credit_card,
                        is_active_member, estimated_salary] + encoded_geography + encoded_gender])

# Feature scaling
user_input_scaled = sc.transform(user_input)

# Make prediction
prediction = model.predict(user_input_scaled)

# Define labels
labels = {1: "Person will leave", 0: "Person will not leave"}

print("Churn Prediction:", labels[int(prediction[0, 0])])
```
